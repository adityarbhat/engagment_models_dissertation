```python
hot_news_entity_ids=pd.read_csv('/Users/adityaravindrabhat/Documents/GitHub/engagement_models/Datasets/MetaAndTextDatasets/hot_news/hot_news_entity.csv')
hot_news_meta=pd.read_csv('/Users/adityaravindrabhat/Documents/GitHub/engagement_models/Datasets/MetaAndTextDatasets/hot_news/hot_news_articles_metamaincombined.csv',parse_dates=True)
hot_news_liwc=pd.read_csv('/Users/adityaravindrabhat/Documents/GitHub/engagement_models/Datasets/MetaAndTextDatasets/hot_news/LIWC_hot_news_body.csv')
hot_news_meta.reset_index(inplace=True)
hot_news_liwc=hot_news_liwc[['id','WC','WPS','Authentic','Tone','Analytic','Clout','posemo','negemo']]
hot_news_meta=hot_news_meta.drop(['title_x','title_y','domain_x','url','author','links','tags','movies','flair','publish_date','created_utc'],axis=1)
print(hot_news_entity_ids.info())
print(hot_news_meta.info())
print(hot_news_liwc.info())
```

```python
hot_news_meta['created_date']=pd.to_datetime(hot_news_meta['created_date'])
hot_news_meta['created_time']=pd.to_datetime(hot_news_meta['created_time'])
hot_news_meta['created_time']=hot_news_meta['created_time'].dt.time
hot_news_meta['day_of_week']=hot_news_meta['created_date'].dt.day_name()
hot_news_meta.info()
```
