```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
from collections import Counter
from spacy import displacy
nlp = spacy.load('en_core_web_md')
```

```python
#Function to extract NER
def entity_extraction(text):
    all_ent_labelst=[]
    doc=nlp(text.strip())
    ents=doc.ents
    all_ent_labelst += [ent.label_ for ent in ents]
    #all_ent_labelst =list(all_ent_labelst)
    c=Counter(all_ent_labelst)
    return c
```

```python
def entity_counter(text):
    all_ent_labelst=[]
    for t in text:
        doc=nlp(t.strip())
        ents=doc.ents
        all_ent_labelst += [ent.label_ for ent in ents]
    c=Counter(all_ent_labelst)
    return c
```

### Hot News NER ANALYSIS

```python
hot_news=pd.read_csv('/Users/adityaravindrabhat/Documents/GitHub/engagement_models/Datasets/MetaAndTextDatasets/hot_news/hot_news_body.csv')
hot_news=hot_news[hot_news['body'].notnull()]
print(hot_news.info())
```

```python
hot_news['entities']=hot_news['body'].apply(entity_extraction)
print(hot_news.info())
print(hot_news.tail())
```

```python
hot_news.reset_index(inplace=True)
```

```python
all_ent_labelst=[]
for t in hot_news['body']:
    doc=nlp(t.strip())
    ents=doc.ents
    all_ent_labelst += [ent.label_ for ent in ents]
c=Counter(all_ent_labelst)
print(c)
```

```python
#Converting the entities column into a dictionary to create a new dataframe with just entity counts
hot_news_dict=dict(hot_news['entities'])
hot_news_entity_count=pd.DataFrame(hot_news_dict)
hot_news_entity_count=hot_news_entity_count.T
hot_news_entity_count=hot_news_entity_count.fillna(0)
print(hot_news_entity_count.info())
print(hot_news_entity_count.tail())
```

```python
hot_news_entity_count.reset_index(inplace=True)
print(hot_news_entity_count.info())
print(hot_news_entity_count.tail())
```

```python
namess=['id','body','entities','index','FAC','DATE','CARDINAL','GPE','ORG','PERSON','TIME','WORK_OF_ART','ORDINAL','LOC','NORP','PRODUCT','MONEY','LAW','EVENT','PERCENT','QUANTITY','LANGUAGE']
len(namess)
```

```python

hot_news_entity_id=pd.concat([hot_news,hot_news_entity_count],axis=1,names=namess)
hot_news_entity_id.drop(['body','entities','index'],axis=1,inplace=True)
#hot_news_entity_id.reset_index(inplace=True)
print(hot_news_entity_id.info())
print(hot_news_entity_id.head())
```

```python
hot_news_entity_id.to_csv('/Users/adityaravindrabhat/Documents/GitHub/engagement_models/Datasets/MetaAndTextDatasets/hot_news/hot_news_entity.csv')
```

```python
#Creating a heatmap to understand the relationship between different entities
hot_news_corr=sns.heatmap(data=hot_news_entity_count.corr(), annot=True, cmap='Spectral',vmin=-1,vmax=1);
```
